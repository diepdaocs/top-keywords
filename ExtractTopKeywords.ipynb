```python
# Clean up output
contents = []
for url in urls:
    crawled_page = url_page_contents.get(url)
    print '############################### Start content for url: %s ################################' % url
    if not crawled_page['error'] and crawled_page['content']:
        content = crawled_page['content']
        content = re.sub(r'[^\w\s]+', '', content)
        content = re.sub(r'(\t|\s)+', ' ', content)
        content = content.lower()
        contents.append(content)
        print content[0:200]
    elif not crawled_page['content']:
        print '- Url "%s" has empty content'
    else:
        print '- Crawl url "%s" error: %s' % (url, crawled_page['error'])
    print '############################### End content for url: %s ################################' % url
```

```python
# Get top keywords
vectorizer = TfidfVectorizer(max_df=0.7, min_df=0.2,max_features=200, stop_words='english')
result = vectorizer.fit_transform(contents)
```

```python
features = vectorizer.get_feature_names()
```

```python
indices = np.argsort(vectorizer.idf_)[::-1]
```

```python
top_n = 20
print '- Top %d keywords' % top_n
top_keywords = [features[i] for i in indices[:top_n]]
for keyword in top_keywords:
    print keyword
```
